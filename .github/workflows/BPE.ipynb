#### Пример расчета прямого распространения сигнала, обратного распространения ошибки и изменения весов сети

```python
#dz=np.array([])
"""""
w, b - синаптические коэффициенты
a- выходы нейронов 
Номера в обозначениях означают следующее 
первая цифра - номер слоя, 
вторая цифра после подчеркивания - номер нейрона 
Например, если на вход сети подается вектор 1 0 1, то a0_1 - значение входа нейронной сети (нулевой или входной слой) 
со значением 0
или для весов 
w1_11 - вес нейрона первого слоя связывюший его со входом a0_1. Этот вес =1
"""""
#
a0_0=1
a0_1=0
a0_2=1
b1_1=1
b1_2=1

a1_0=1
b2_1=1

w1_11=1
w1_12=0.5
w1_21=-1
w1_22=2
w2_11=1.5
w2_12=-1

ro=0.5 #learning rate
```

```python
#Forward propagation #1
z1_1=b1_1 + w1_11*a0_1 +w1_12 *a0_2 
z1_2=b1_2 + w1_21*a0_1 +w1_22 *a0_2
z1_2
a1_1=1/(1+np.exp(-z1_1))
a1_2=1/(1+np.exp(-z1_2))
print(a1_1)
print(a1_2)
z2_1=b2_1+w2_11*a1_1+w2_12*a1_2
a2_1=1/(1+np.exp(-z2_1))
print(a2_1)
y=1 #target value
print('Error=', y-a2_1)
```

```python
#Back propagation #1
#a2=0.78139
#w
dz2_1=y-a2_1
dz1_1=dz2_1*w2_11*a1_1*(1-a1_1)
dz1_2=dz2_1*w2_12*a1_2*(1-a1_2)
print(dz1_1)
print(dz1_2)
w2_11=w2_11+ro*dz2_1*a1_1
w2_12=w2_12+ro*dz2_1*a1_2
b2_1=b2_1+ro*dz2_1*a1_0
w1_11=w1_11+ro*dz1_1*a0_1
w1_12=w1_12+ro*dz1_1*a0_2
w1_21=w1_21+ro*dz1_2*a0_1
w1_22=w1_22+ro*dz1_2*a0_2
b1_1=b1_1+ro*dz1_1*a0_0
b1_2=b1_2+ro*dz1_2*a0_0
print('Corrected weigths')
print(b1_1)
print(b1_2)
print(w1_11)
print(w1_12)
print(w1_21)
print(w1_22)
print(b2_1)
print(w2_11)
print(w2_12)
```
